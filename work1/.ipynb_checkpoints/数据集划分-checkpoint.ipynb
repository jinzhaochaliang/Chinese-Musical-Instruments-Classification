```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_excel("data.xlsx")
```

```python
n = df.shape[0]
d = {}
for i in range(n):
    if str(df.iloc[i,1]) in d:
        d[str(df.iloc[i,1])] += int(df.iloc[i,5])
    else:
        d[str(df.iloc[i,1])] = int(df.iloc[i,5])
print(str(d).replace('\'','\"'))
print(d)
```

```python
v = d.values()
v1 = []
for i in v:
    if i<10:
        v1.append(int(i*0.5))
    else:
        v1.append(int(i*0.5))
print(v)
print(v1)
print(sum(v1))

v2 = []
count = 0
for i in v:
    v2.append(count)
    count += i
print(v2)

v3 = []
for i,j in zip(v2,v1):
    v3.append(i+j)
print(v3)

v4 = []
count = 0
for i in v:
    count += i
    v4.append(count) 
print(v4)
```

```python
train = pd.DataFrame(columns = ['name','label','cut','manually_verification','d']) 
test = pd.DataFrame(columns = ['name','label','cut']) 
```
